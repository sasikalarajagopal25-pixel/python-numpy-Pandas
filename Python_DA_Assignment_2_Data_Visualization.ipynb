# **Visualizations using Matplotlib/Pandas Plot**

# **Line Chart**

Plot a line chart to visualize the fare over time, using the pickup timestamp as the
x-axis and fare as the y-axis. Ensure the pickup column is converted to a datetime
format before plotting.

```python
df['pickup'] = pd.to_datetime(df['pickup'])
df = df.sort_values('pickup')
plt.figure(figsize=(10, 5))
plt.plot(df['pickup'], df['fare'], color='royalblue', linestyle='-', linewidth=2.5, label='Fare')
plt.title('Fare Over Time', fontsize=15, fontweight='bold', color='darkblue', pad=15)
plt.xlabel('Pickup Time', fontsize=12, color='darkblue')
plt.ylabel('Fare Amount', fontsize=12, color='darkblue')
plt.legend(frameon=True, facecolor='white')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

# **Bar Chart**

```python
fare_total=df['fare'].sum()
fare_total
```

```python
pickup_borough=df.groupby('pickup_borough')['fare'].sum()
pickup_borough
```

Create a bar chart to show the total fare for each pickup_borough. Group the data
by pickup_borough and sum the fare for each group.

```python

fare_by_borough = df.groupby('pickup_borough')['fare'].sum().reset_index()
plt.figure(figsize=(8,5))
plt.bar(fare_by_borough['pickup_borough'], fare_by_borough['fare'],
        color='royalblue', edgecolor='black', width=0.6)
plt.title('Total Fare by Pickup Borough', fontsize=15, fontweight='bold', color='darkblue')
plt.xlabel('Pickup Borough', fontsize=12, color='black')
plt.ylabel('Total Fare', fontsize=12, color='black')
plt.grid(axis='y', linestyle=':', color='gray', alpha=0.7)
plt.tight_layout()
plt.show()
```

# **Pie Chart**

Plot a pie chart showing the distribution of trips based on the payment method (credit
card, cash, etc.). Each slice should represent the count of trips for a specific payment
method.

```python
df.groupby('payment')['payment'].count().plot(kind='pie',autopct='%1.0f%%',startangle=140,figsize=(8,5))
plt.title('Distribution of trips based on the payment',fontsize=12,color='black')
plt.ylabel('')
plt.show()
```

# **Histogram**

Create a histogram to visualize the distribution of distance. Customize the number of
bins for better granularity and ensure the plot is easy to interpret.

```python
plt.figure(figsize=(8,5))
plt.hist(df['distance'],bins=20,color='royalblue',edgecolor='black')
plt.title('Distribution of distance',fontsize=12)
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Box Plot**

Plot a box plot to visualize the distribution of tip amounts for each pickup_borough.
Use pickup_borough as the categorical axis and tip as the numeric axis.

```python
plt.figure(figsize=(8,5))
df.boxplot(column='tip',by='pickup_borough',grid=False)
plt.title('Distribution of tip amount for each pickup borough',fontsize=12)
plt.xlabel('Pickup Borough')
plt.ylabel('Tip Amount')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

# **Visualizations using Seaborn**

# **Count Plot**

Create a count plot to visualize the number of trips in each pickup_borough. The
x-axis should represent the boroughs, and the y-axis should show the count of trips.

```python
ax=sns.countplot(data=df,x='pickup_borough',color='royalblue')
ax.bar_label(ax.containers[0])
plt.title('Number of trips in each pickup borough')
plt.show()
```

# **Scatter Plot**

Plot a scatter plot to show the relationship between distance and fare. Use
distance on the x-axis and fare on the y-axis to visualize any correlation. Color the
points based on the pickup_borough to differentiate the trips by their respective
boroughs.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='distance',y='fare',hue='pickup_borough',palette='bright',s=100)
plt.title('Relationship between distance and fare')
plt.show()
```

# **Heatmap**

Plot a heatmap to visualize the correlation between numerical variables such as
distance, fare, tip, tolls, and total. Use a correlation matrix to highlight the
relationships.

```python
correlation_matrix=df[['distance','fare','tip','tolls','total']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation between numerical variables')
plt.yticks(rotation=45)
plt.show()
```

# **Pair Plot**

Create a pair plot to visualize the pairwise relationships between distance, fare, tip,
and total. Color the data points according to the pickup_zone method to compare
how different zones affect these variables.

```python
correlation_matrix=df[['distance','fare','tip','total']]
correlation_matrix
```

```python
plt.figure(figsize=(10, 6))
sns.pairplot(data=df, vars=correlation_matrix,hue='pickup_zone')
plt.suptitle('')
plt.show()
```

# **Violin Plot**

Plot a violin plot to show the distribution of fare for each payment method. Use the
payment method as the categorical axis and fare as the numeric axis to visualize its

```python
sns.violinplot(data=df,x='payment',y='fare',color='royalblue')
plt.title('Distribution of fare for each payment')
plt.show()
```
