# **Pandas Series**

# **Creating Pandas Series**

```python
import pandas as pd
```

```python
Mark=[95, 92, 89, 85, 80]
Custom_Indices=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
marks = pd.Series(data=Mark,index=Custom_Indices)
print(marks)
```

# **Indexing and Slicing**

Use the integer index position to access the mark of the 1st rank student.

```python
first_rank_mark = marks.iloc[0]

print("Mark of 1st rank student:", first_rank_mark)
```

Use the loc accessor to retrieve the marks of the top 3 ranks by
specifying their index labels.

```python
top3_rank_mark = marks.loc['Rank1': 'Rank3']

print("Mark of 1st rank student:\n", top3_rank_mark)
```

Use the iloc accessor to retrieve the mark of the 3rd rank student.

```python
third_rank_mark = marks.iloc[2]
print("3rd_rank_mark:",third_rank_mark)
```

Apply a boolean mask to filter the Series and retrieve the ranks where the
marks are greater than 90.

```python
greater_than_90= marks[marks >90]
print("Greater than 90 marks:\n",greater_than_90)
```

```python
print("boolean filter in greater than 90:\n",marks>90)
```
